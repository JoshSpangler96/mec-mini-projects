## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

```python
# load the raw json data to handle nested fields
json_file = open('data/world_bank_projects.json')
json_raw = json.load(json_file)
json_file.close()
```

```python
# Normalize the data with json_raw and get a data frame for mjtheme_namecode
project_themes_df = json_normalize(json_raw, 'mjtheme_namecode').groupby('name').count().rename({'code':'count'}, axis=1)
project_themes_df = project_themes_df.reset_index().sort_values('count', ascending = False)

# Drop blank names 
project_themes_df = project_themes_df.replace(r'^\s*$', np.nan, regex=True).dropna()

print('Top 10 project Themes:')
project_themes_df.head(10)
```

In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
# Create a new normalized json
name_themes_df = json_normalize(json_raw, 'mjtheme_namecode')
name_themes_df.drop_duplicates

# drop all empty string or na values 
name_themes_df = name_themes_df.replace(r'^\s*$', np.nan, regex=True).dropna()

# Associate each code to a unique value
name_themes_df = pd.DataFrame(name_themes_df.groupby('code')['name'].unique())
name_themes_df['name'] = [x[0] for x in name_themes_df['name']]
name_themes_df.reset_index()
```

```python
# load the mjtheme_namecode into a new normalized json dataframe
themes_df = json_normalize(json_raw, 'mjtheme_namecode')
themes_df.head(20)
```

```python
themes_df.shape
```

```python
# merge the name_themes_df with themes_df to associate each code to the project theme
themes_clean_df = themes_df.merge(name_themes_df, how='outer', left_on=['code'], right_index=True)
themes_clean_df
themes_clean_df = pd.DataFrame(themes_clean_df[['name_y', 'code']])
themes_clean_df.rename({'name_y': 'project_theme'}, axis=1)
themes_clean_df.sort_index().head(20)
```

```python
# Verify that there are no empty strings
themes_clean_df.replace(r'^\s*$', np.nan, regex=True)
themes_clean_df.isna().sum()
```
